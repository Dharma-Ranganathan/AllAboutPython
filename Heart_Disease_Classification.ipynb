Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

Data collection

```python
heart_disease = pd.read_csv('/content/drive/MyDrive/Colab_python/heart_disease_data.csv')
heart_disease.head()
```

Data Pre-processing

```python
# checking if null values exists

heart_disease.info()
```

```python
# rechecking
heart_disease.isnull().sum()
```

* Null values doesn't exists in our dataset.

```python
# basic statistical report
heart_disease.describe()
```

```python
# checking shape of dataset
heart_disease.shape
```

```python
# checking if dataset is in balanced state
s1 = heart_disease['target'] == 1
s2 = heart_disease['target'] == 0
print(s1.sum(),s2.sum(),sep='  ')
```

```python
# one more approach in one line as same as previous
heart_disease.value_counts('target')
```


* dataset in balanced state only
* splitting features and label seperately

```python
feature = heart_disease.drop('target',axis=1)
label = heart_disease['target']
```

* drops 'target' column and axis 1 says it's belongs to column
* Enabling target column to label variable itself

```python
# checking feature
feature.head()
```

```python
# checking label
label.head()
```

Model Selection

```python
x_train,x_test,y_train,y_test = train_test_split(feature,label,test_size=0.2, stratify=label,random_state=42)
```

```python
# checking how many test datas were taken
print(feature.shape,x_train.shape,x_test.shape)
```

* 242 datas for training
* 61 datas for testing

Model Training

```python
# model - binary classification
# Logistic regression

model = LogisticRegression()
```

```python
# fitting training dataset to model

model.fit(x_train,y_train)
```

Model Evaluation and it's Reports

```python
# prediction of seen data - training data

x_train_pred = model.predict(x_train)

# accuracy

train_score = accuracy_score(x_train_pred,y_train)

print(f'accuracy of X trained : {train_score:.2f}')
print(f'percentage : {(train_score * 100):.2f}%')
print(f'out of {x_train.shape[0]} trained data, {(train_score * x_train.shape[0]):.0f} were accurate')
```

```python
# prediction of unseen data - test data

x_test_pred = model.predict(x_test)

# accuracy

test_score = accuracy_score(x_test_pred,y_test)

print(f'accuracy of X test : {test_score:.2f}')
print(f'percentage : {(test_score * 100):.2f}%')
print(f'out of {x_test.shape[0]} test data, {(test_score * x_test.shape[0]):.0f} were accurate')
```

Classification reports

```python
# reports of unseen data

test_reports = classification_report(x_test_pred,y_test)

print(test_reports)
```
