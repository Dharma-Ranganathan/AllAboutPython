1. Odd or Even

```python
num = int(input('enter num to check odd or even : '))

if num%2 != 0:
    print(f'given {num} is odd')
else:
    print(f'given {num} is even')
```

2. Negative | Positive | Zero

```python
num = int(input())

if num > 0:
    print(f'given {num} is positive')
elif num < 0:
    print(f'given {num} is negative')
elif num == 0:
    print(f'given {num} is zero')
```

3. Factorial of num

```python
def fact(num):
    if num == 0:
        return 1
    return num * fact(num-1)

num = int(input())
val = fact(num)
print(f'factorial of {num} is {val}')
```

4. Sum of digits

```python
num = input().strip()

# function
def parse_sum(num):
    return sum(list(map(int,num)))

if num.isnumeric():
    val = parse_sum(num)
    print(f'sum of digits of {num} is {val}')
else:
    print('invalid credentials')
```

5. Reverse a number

```python
num = input().strip()

if num.isnumeric():
    print(f'{num} reversed as {num[::-1]}')
else:
    print('invalid credentials')
```

6. Check palindrome for num

```python
num = input().strip()

if num.isnumeric():
    if num == num[::-1]:
        print(f'given {num} is palindrome')
    else:
        print(f'given {num} is not palindrome')
else:
    print('invalid credentials')
```

7. Greatest of three numbers

```python
num = list(map(int,input('enter numbers :').split()))

max = 0
for n in num:
    if n > max:
        max = n

print(f'greatest number in {num} is {max}')
```

8. Check prime

```python
num = int(input())

def check_prime(num):
    count = 0
    for i in range(1,num+1):
        if num%i == 0:
            count += 1
    return count


count = check_prime(num)

if count == 2:
    print(f'given {num} is prime')
else:
    print(f'given {num} is not prime')
```

11. Power by loops

```python
a = int(input())
b = int(input())

power = 1
for i in range(b):
    power = a*a

print(f'power {a} by {b} is {power}')
```

12. Fibonacci sequence

```python
n = int(input())
print('series of Fibonacci')
print()
n1 = 0
n2 = 1
print(n1,end=' • ')
print(n2,end=' • ')
for i in range(n+1):
    n3 = n2+n1
    print(n3,end=' • ')
    n1 = n2
    n2 = n3
```

17. LCM of two numbers

```python
# function for gcd
def gcd(a, b):
    return a if b == 0 else gcd(b, a % b)

def lcm(a, b):
    return (a // gcd(a, b)) * b


a = 10
b = 5
print(lcm(a, b))
```
