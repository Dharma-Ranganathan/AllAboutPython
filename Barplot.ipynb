Barplot using matplotlib

```python
# import matplotlib
import matplotlib.pyplot as plt
```

```python
# import numpy
import numpy as np
```

```python
# dataset - custom
x = np.linspace(1,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
# creating bar plot
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(x,y,color='#ff00ce')
ax.bar(x,z,color='#00ffce')
plt.xlabel('angles(theta)')
plt.ylabel('sine and cosine')
plt.title('sine vs cosine')
plt.show()
```
