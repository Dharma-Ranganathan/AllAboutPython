Custom scatter plot

Importing required libraries

```python
# numpy
import numpy as np
```

```python
# matplotlib
import matplotlib.pyplot as plt
```

Creating custom datasets using numpy

```python
# using linspace for evenly spaced values
x = np.linspace(1,10,100)
```

Creating sine and cosine values of x

```python
# sine values
y = np.sin(x)

# cosine values
z = np.cos(x)
```

Creating scatter plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.scatter(x,y,c='r',marker='o')
ax.scatter(x,z,c='g',marker='x')
plt.xlabel('angles(theta)')
plt.ylabel('sine and cosine')
plt.title('sine vs cosine')
plt.show()
```
