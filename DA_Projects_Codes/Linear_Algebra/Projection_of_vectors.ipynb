**Projection of two vectors**

**Formula :**

(dot product of two vectors AB / magnitude of B vetor) * vector B

**Dot product of two vectors AB :**

--> A â€¢ B

**Magnitude of vector B squared:**

--> Square root of square of vector B1 plus square of vector B2

--> Atlast multiplied by vector B

```python
# libraries

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# two vectors

vect_a = np.array([3,4])

vect_b = np.array([8,9])
```

```python
# seaborn theme
sns.set_theme()
```

```python

# visualise vector before projection

plt.quiver(0,0,3,4,scale_units='xy',angles='xy',color='b',scale=1)
plt.quiver(0,0,8,9,scale_units='xy',angles='xy',color='y',scale=1)
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

```python
# apply projection

numerator = np.dot(vect_a,vect_b)

denominator= np.sqrt(sum(np.square(vect_b)))

multiplier = vect_b

projection_of_vectors = (numerator/denominator**2)*multiplier

#projection_of_vectors = np.array(projection_of_vectors,dtype=int)

print(projection_of_vectors)
```

```python

# visualise vector after projection

plt.quiver(0,0,3,4,scale_units='xy',angles='xy',color='b',scale=1)
plt.quiver(0,0,8,9,scale_units='xy',angles='xy',color='y',scale=1)
plt.quiver(0,0,3.310,3.724,scale_units='xy',angles='xy',color='r',scale=1)
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```
