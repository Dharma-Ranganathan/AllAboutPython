Linear Algebra - vectors

```python
# libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Vectors and Scalar**

Vector - collection of values in either single row or single column

col - [[12 24 36]],
row - [[12],[24],[36]]

We say vector contains both magnitude and direction

Magnitude - values in it
Direction - rows and columns

Scalar - simply a magnitude itself

Any value without square brackets

val - 45 - scalar value

```python
# vector addition - two vectors

vect_a = np.array([5,4])

vect_b = np.array([1,2])

vect_a_sum_b = vect_a + vect_b

print(vect_a_sum_b)
```

Using numpy methods

```python
sum = np.add(vect_a,vect_b)

print(sum)
```

```python
# visualise vector addition

# before
sns.set()
plt.quiver(0,0,5,4,scale_units='xy',angles='xy',color='b',scale=1)
plt.quiver(0,0,1,2,scale_units='xy',angles='xy',color='y',scale=1)
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

```python
# visualise vector addition

# after
plt.quiver(0,0,5,4,scale_units='xy',angles='xy',color='b',scale=1)
plt.quiver(0,0,1,2,scale_units='xy',angles='xy',color='y',scale=1)
plt.quiver(0,0,6,6,scale_units='xy',angles='xy',color='r',scale=1)
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

```python
# vector subtraction - two vectors

vect_c = np.array([5,6])

vect_d = np.array([8,9])

vect_c_sub_d = vect_c - vect_d

print(vect_c_sub_d)
```

Using numpy methods

```python
diff = np.subtract(vect_c,vect_d)

print(diff)
```

```python
# visualise vector subtraction

# before
plt.quiver(0,0,5,6,scale_units='xy',angles='xy',color='b',scale=1)
plt.quiver(0,0,8,9,scale_units='xy',angles='xy',color='y',scale=1)
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

```python

# visualise vector subtraction

# after
plt.quiver(0,0,5,6,scale_units='xy',angles='xy',color='b',scale=1)
plt.quiver(0,0,8,9,scale_units='xy',angles='xy',color='y',scale=1)
plt.quiver(0,0,-3,-3,scale_units='xy',angles='xy',color='r',scale=1)
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

Multiplication - vector with scalar

```python

# scalar

scal_a = 2

# vector

vect_e = np.array([2,4])

# scalar * vector = vector only

vect_mul = vect_e * scal_a

print(vect_mul)
```

```python

# visualise multiplication vector with scalar
# simply increase the vector size

# before
plt.quiver(0,0,2,4,scale_units='xy',angles='xy',color='b',scale=1)
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

Using numpy methods

```python
mul = np.multiply(vect_e,scal_a)

print(mul)
```

```python

# visualise multiplication vector with scalar
# simply increase the vector size

# after
plt.quiver(0,0,2,4,scale_units='xy',angles='xy',color='b',scale=1)
plt.quiver(0,0,4,8,scale_units='xy',angles='xy',color='y',scale=1)
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

Multiplication - vector with vector

```python
# vector

vect_f = np.arange(1,3,1)

# vector

vect_g = np.arange(3,5,1)

vects_mul = vect_f * vect_g

print(vects_mul)
```

Using numpy methods

```python
mul_2 = np.multiply(vect_f,vect_g)

print(mul_2)
```

Dot product - two vectors - results in scalar

```python

# two vectors to scalar

dot_1 = np.random.randint(1,100,2)
dot_2 = np.random.randint(1,100,2)

dot_1_dot_dot_2 = np.dot(dot_1,dot_2)

print(dot_1_dot_dot_2)
```

Cross product - two vectors - results in vector only

```python

# two vectors to vector only

cross_1 = np.random.randint(1,100,3)
cross_2 = np.random.randint(1,100,3)

cross_1_cross_cross_2 = np.cross(cross_1,cross_2)

print(cross_1_cross_cross_2)
```
