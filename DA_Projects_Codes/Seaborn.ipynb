Seaborn basics

```python
# importing seaborn
import seaborn as sns

# import matplotlib
import matplotlib.pyplot as plt
```

Handling in-built datasets

```python
# calling iris using seaborn
iris = sns.load_dataset('iris')
iris.head()
```

```python
# scatter plot
sns.set_theme()
sns.relplot(data=iris,x='sepal_width',y='sepal_length',hue='species',style='species',size='species',kind='scatter',row='species')

plt.show()
```

```python
# Titanic dataset
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# countplot
sns.countplot(x='alive',data=titanic,hue='class')
```

```python
# values of each classes
titanic.value_counts('alive')
```

Distribution plot

```python
sns.histplot(titanic['survived'],kde=True)
```
