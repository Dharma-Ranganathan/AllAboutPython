Heatmap using seaborn

```python
# imports
import seaborn as sns
import matplotlib.pyplot as plt

# dataset
iris = sns.load_dataset('iris')
del iris['species']

# correlation of iris dataset
mapp = iris.corr()

# plotting and visualise
plt.xlabel(None)
plt.ylabel(None)
plt.title('heat map')
sns.heatmap(mapp,xticklabels=False,yticklabels=True,cbar=True,square=True,cmap='Greens', linewidths=2, linecolor='white')
plt.show()
```
