#**EDA & Data Visualization**

**Dependencies**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Data Collection**

```python
breast_cancer = pd.read_csv('/content/drive/MyDrive/Colab_python/breast_cancer_data.csv')

# check
breast_cancer.head()
```

**Get rid of unwanted columns**

```python
# dropping unwanted columns

breast_cancer = breast_cancer.drop(['id','Unnamed: 32'],axis=1)

# check
breast_cancer.head()
```

**Checking Null**

```python
# shape
breast_cancer.shape
```

```python
# check null

breast_cancer.info()
```

```python
# checking null and sum

breast_cancer.isnull().sum()
```

**Check Imbalanced State**

```python
# check balance state

breast_cancer.value_counts('diagnosis')
```

* more or less , our dataset is in balanced state

**Splitting features and label**

```python
features = breast_cancer.drop('diagnosis',axis=1)
label = breast_cancer['diagnosis']
```

```python
# check features
features.head()
```

```python
# check label
label.head()
```

**Label Encoding Manual**

```python
# B - 0 - normal
# M - 1 - defective

label = np.where(label == 'B',0,1)

# check label
print(label)
```

**Descriptive Statistics**

```python
features.describe()
```

**Stats and Visual**

```python
for column in features:
    print(column)
```

```python
for column in features:
    print(features[column].describe())
    print()
    print('----------------------------------')
```

* if mean > median : right skewed
* if median > mean : left skewed

* Skewed - வளைந்த

```python
for column in features:
    sns.histplot(x=column,data=features,kde=True)
    plt.show()
```

**Label Visual**

```python
# visualization of label
sns.countplot(x='diagnosis', data = breast_cancer,hue='diagnosis')
```

```python
# counts the target
breast_cancer.value_counts('diagnosis')
```

**Inference: 'based on the eda we have done**

* No null values exists
* More or less in balanced state
* More benign , less malignant
* Benign - normal human
* Malignant - affected human

**Descriptive Statistics**

* Able to notice , mean > median
* Right positive skewness
* If null values exists, we can use mean itself
* Data Visualization - Right skewness occured using hist plot
* We have to do Standardization methods in data points

```python
# group by the diagnosis

breast_cancer.groupby('diagnosis').mean()
```
