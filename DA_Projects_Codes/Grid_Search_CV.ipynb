Grid search cross validation is used to find hyperparameters of a model by tuning it.

Dependencies

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import GridSearchCV
```

Data collection

```python
breast_cancer = pd.read_csv('/content/drive/MyDrive/Colab_python/breast_cancer_data.csv')

# checking
breast_cancer.head()
```

```python
print(breast_cancer)
```

```python
breast_cancer.tail()
```

Something at last column is NaN , we need to drop that.

```python
# dropping last column and id column
breast_cancer = breast_cancer.dropna(axis=1)
breast_cancer = breast_cancer.drop('id',axis=1)

# checking
breast_cancer.head()
```

Data Pre-processing

```python
# checking null values
breast_cancer.isnull().sum()
```

* no null values apart from last column we dropped

```python
# checking shape
breast_cancer.shape
```

```python
# checking balanced state
breast_cancer.value_counts('diagnosis')
```

```python

# benign

benign = breast_cancer['diagnosis'] == 'B'
benign = benign.sum()
benign
```

```python
# malignant

malignant = breast_cancer['diagnosis'] == 'M'
malignant = malignant.sum()
malignant
```

* Total data points = 569
* Benign data points = 357
* Malignant data points = 212

```python
# checking proportionality of diagnosis columns

benign_prop = (benign / breast_cancer.shape[0]) * 100
malignant_prop = (malignant/breast_cancer.shape[0]) * 100

print(round(benign_prop,2),'%', round(malignant_prop,2),'%')
```

* our dataset is not in that much Imbalanced, more or less is near to Imbalanced but no problem. We can proceed

Splitting features and labels(diagnosis)

```python
features = breast_cancer.drop('diagnosis',axis=1)
label = breast_cancer.diagnosis

print('features')
print(features)
print()

print('label')
print(label)
print()
```

Label Encoding using LabelEncoder class

```python
# label encoding
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
label = encoder.fit_transform(label)
print(label)
```

Converting features and label into numpy arrays for grid search cross validation

```python
# as numpy arrays
x = np.asarray(features.values)
y = np.asarray(label)

print(x)
print()
print(y)
```

Implementation of grid search cv

```python
# creating param_grid for grid search cv
params = { 'kernel' : ['kernel','poly','rbf','sigmoid'], 'C' : [1,5,10,20] }
```

```python
# model svm
from sklearn.svm import SVC
model = SVC()
```

```python
classifier = GridSearchCV(model,params,cv=5)
```

Providing x and y in classifier

```python
classifier.fit(x,y)
```

```python
# results of classifier
classifier.cv_results_
```

```python
# best params we got

best_params = classifier.best_params_
print(best_params)
```

```python
# cv results to dataframe

results = pd.DataFrame(classifier.cv_results_)
results.head()
```

Creating a new data frame out of this results data frame

```python
insight = results[['param_C','param_kernel','mean_test_score','rank_test_score']]

print(insight)
print()
```

We got to know , rank 1 has kernel 'rbf' and C=5

Using this hyperparameters, we can develop a new model
