Heatmap using seaborn

```python
# imports
import seaborn as sns
import matplotlib.pyplot as plt

# dataset
iris = sns.load_dataset('iris')
del iris['species']
# correlation of iris dataset
map = iris.corr()

# plotting and visualise
plt.xlabel(None)
plt.ylabel('')
plt.title('heat map')
sns.set_style('dark')
sns.heatmap(map,xticklabels=False,cbar=False,square=True,cmap='Greens')
plt.show()
```
