3d scatter plot

```python
# import matplotlib
import matplotlib.pyplot as plt

# import numpy
import numpy as np
```

```python
# custom - datasets
x = np.arange(1,100,1)
y = np.sin(x)
z = np.cos(x)
```

```python
# plotting and visualise
ax = plt.axes(projection = '3d')
ax.scatter(x,y,z,c=x,cmap='Spectral')
plt.title('3d sine vs cosine')
plt.show()
```
