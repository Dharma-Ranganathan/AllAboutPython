Use case - Converts textual data into numerical data through vectorization

```python
# importing required libraries
import seaborn as sns
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# datasets
iris = sns.load_dataset('iris')
iris.head()
```

```python
# getting textual data from iris

texts = iris.species

print(texts)
```

```python
# inserting texts to TfidfVectorizer

vectorizer = TfidfVectorizer()

vectorizer.fit(texts)

texts = vectorizer.transform(texts)

print(texts) # results 1.0 - cause, only one time that text arises on each search
```
