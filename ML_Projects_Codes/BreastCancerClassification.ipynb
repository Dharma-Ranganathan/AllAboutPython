#**Model Selection**

```python
# cross validation score for choosing right model with default parameters

# cross models list
cross_models = [ LogisticRegression(max_iter=100), SVC(kernel='linear'), KNeighborsClassifier(n_neighbors=3), RandomForestClassifier(n_estimators=100)]
```

**Cross Validation score**

```python
def compare_model_cross_val(models,x,y):
    for model in models:
        score = cross_val_score(model,x,y,cv=5)

        # mean accuracy
        accuracy = sum(score)/len(score)
        accuracy = round(accuracy*100,2)

        # check
        print('----------------------------------------------------------')
        print(f'scores of each {model} : {score}')
        print(f'percentage of each {model} : {accuracy}%')
        print('----------------------------------------------------------')
```

**GridSearchCV**

```python
hyperparameter = {
    'log_reg': {
        'max_iter':[50,100,500,1000]
    },
    'svc' : {
        'kernel':['linear','poly','sigmoid','rbf'],
        'C': [1,3,5,10]
    },
    'knn' : {
        'n_neighbors':[3,5,7,10]
    },
    'rfc' : {
        'n_estimators':[50,100,200,300]
    }
}
```

```python
model_keys = list(hyperparameter.keys())
model_keys
```

```python
# new instances of models
grid_models = [LogisticRegression(),SVC(), KNeighborsClassifier(), RandomForestClassifier()]
```

```python
def compare_model_grid_cv(models,x,y):
    result = []
    i = 0
    for model in models:
        params = hyperparameter[model_keys[i]]
        i += 1

        grid = GridSearchCV(model,params,cv=5)
        grid.fit(x,y)

        result.append({
            'score':grid.best_score_,
            'model':model,
            'params':grid.best_params_
        })
        return result
```

**Split Features and label**

```python
features = breast_cancer.drop('target',axis=1)
label = breast_cancer.target

# check
features.head()
```

```python
# check
label.head()
```

```python
# cross validation score

compare_model_cross_val(cross_models, features.values,label)
```

* In cross validation score, all four models shows good means accuracy above 90%

```python
# grid search cv
result = compare_model_grid_cv(grid_models, features.values,label)

result = pd.DataFrame(result)

result
```

* we may notice, logistic regression model shows score of 95 above, hyperparameters of max_iter is 500

#**Model Building**

**Model Training**

```python
# new instance of logistic regression
model = LogisticRegression(max_iter=500)
```

**Train and Test Data**

```python
x_train,x_test,y_train,y_test = train_test_split(features.values,label,test_size=0.2,stratify=label,random_state=42)
```

```python
model.fit(x_train,y_train)
```

**Accuracy of Seen Data**

```python
train = model.predict(x_train)

# accuracy

train_acc = accuracy_score(y_train,train)
train_acc = round(train_acc *100,2)
print(f'accuracy on seen data : {train_acc}%')
```

**Accuracy of unseen data**

```python
test = model.predict(x_test)

# accuracy

test_acc = accuracy_score(y_test,test)
test_acc = round(test_acc *100,2)
print(f'accuracy on unseen data : {test_acc}%')
```

* so far we have done great progress by making accurate predictions on seen and unseen datum

**Confusion Matrix for Precision and Recall**

```python
cm = confusion_matrix(y_test,test)

# heatmap
sns.heatmap(cm,annot=True,fmt='d',cmap='Greens')
plt.title('confusion matrix')
plt.xlabel('predicted')
plt.ylabel('true values')
plt.show()
```

**Classification Reports**

```python
cr = classification_report(y_test,test)

# check cr
print(cr)
```
