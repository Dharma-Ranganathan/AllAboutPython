Cross Validation Score is to find out the proper and perfect for model selection instead train test split.

Import dependencies

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score
```

Import All classification models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

Data collection for classification model - using heart disease dataset

```python
heart = pd.read_csv('/content/drive/MyDrive/Colab_python/heart_disease_data.csv')

# checking
heart.head()
```

Data Pre-processing

```python

# checking if missing values exists

heart.isnull().sum()
```

* no null values exists

```python
# checking if dataset is in balanced state

heart.shape
```

```python
heart.value_counts('target')
```

* dataset is in balanced state

Basic EDA

```python
# mean by target
heart.groupby('target').mean()
```

```python
# median by target
heart.groupby('target').median()
```

```python
# max and min
Max = heart.groupby('target').max()
print('maximum of each feature by target :')
print(Max)
print()

Min = heart.groupby('target').min()
print('minimum of each feature by target :')
print(Min)
print()
```

Splitting features and Label

```python
features = heart.drop('target',axis=1)
label = heart['target']
```

Train test split by first

```python
x_train,x_test,y_train,y_test = train_test_split(features.values, label, test_size = 0.2, stratify=label,random_state=3)
```

```python
print(features.shape,x_train.shape,x_test.shape)
```

* In 303 data points, 242 is for training and 61 is for testing

Model selection by train_test_split

```python
models = [LogisticRegression(max_iter=1000), SVC(kernel='linear'), KNeighborsClassifier(), RandomForestClassifier()]
```

```python
def compare_score_train_test(models):

    for model in models:

        model.fit(x_train,y_train)

        pred = model.predict(x_test)

        accuracy = accuracy_score(pred,y_test)

        accuracy = round(accuracy,2)

        print(f'Accuracy of {model} is {accuracy}%')
```

```python
# function call

compare_score_train_test(models)
```

* In train test split method way, we got to know logistic model performace and accuracy is higher compared than rest of models

Model selection by cross_val_score instead of train_test method

```python
# getting new instances of models again

cross_models = [
LogisticRegression(max_iter=1000),
SVC(kernel='linear'),
KNeighborsClassifier(),
RandomForestClassifier()
]
```

```python
# function for comparison by cv score

def compare_score_cross_validation(models):
    for model in models:
        score = cross_val_score(model,features.values,label,cv=5)

        # scores
        print(f'scores of {model} :')
        print(score)
        print()

        # mean accuracy
        accuracy = sum(score) / len(score)
        accuracy = round(accuracy,3)
        print(f'accuracy of {model} = {accuracy}%')
        print('---------------------------------------------------------------')
```

```python
# function call
compare_score_cross_validation(cross_models)
```

* we got to know, out of 4 models, 3 models is providing same accuracy values, which means we can able to use those models in our model selection.

Note :
* Cross Validation Score is basically used only on small dataset
* Using this score , we can able to find perfect and efficient model in our training time
* More effective than train test split , but takes more time in large dataset compared to train test split method
* Using of this method could be arranged in needed only way
